# Setup environment and collect initial data

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
```

### Dynamics
This is the underneath dynamics in the mountain car. For this exercise, we will use it, but for the following sessions, **we will learn it!**

```python
def mountain_car_dynamics(p, v, a):
    v_next = v + 0.001 * a - 0.0025 * np.cos(3 * p)
    v_next = np.clip(v_next, -0.07, 0.07)
    p_next = p + v_next
    if p_next < -1.2:
        p_next, v_next = -1.2, 0.0
    return np.array([p_next, v_next])

# Test dynamics, at position p, and velocity v, predict the new location when the action a=1 is applied.
mountain_car_dynamics(p=-0.5, v=0, a=1.0)
```

### Cost function

```python
def cost_fn(state, a):
    x, v = state
    goal = 0.5
    target = goal - x

    # Encourage moving in the right direction and gaining kinetic energy
    cost = target**2 - 1000 * np.abs(v)
    return cost
```

## Roll out trajectories based on action sequences U

```python
def roll_out_trajectory(s0, Ui, gamma=1.0):
    s = s0.copy()  # Initial state
    traj = [s0.copy()]  # Trajectory
    H = len(Ui)
    discount = 1.0

    total_cost = 0.0

    # For each time in the planning horizon
    for k in range(H):
        a = Ui[k]  # continuous action in [-1, 1]
        delta_m = np.zeros(2)

        # Predict Œîx using both GPs (for Œîp, Œîv)
        s = mountain_car_dynamics(s[0], s[1], a)
        traj.append(s.copy())

        # Accumulate discounted cost
        total_cost += cost_fn(s, a) * discount
        discount *= gamma
    return np.array(traj), total_cost


def roll_out_n_trajectories(state, U):
    N_seq = len(U)
    J = np.zeros(N_seq)

    # Roll out trajectories and computer their cost
    all_traj = []
    for i in range(N_seq):
        traj, total_cost = roll_out_trajectory(state, U[i])

        J[i] = total_cost
        all_traj.append(np.array(traj))

    return all_traj, J

# Action squence
Ui = np.zeros(3)
traj, J = roll_out_trajectory(np.array([-0.5,0]), Ui)
traj
```

```python
def plot_trajectories(state, best_traj):
    plt.figure(figsize=(8, 6))
    for i, traj in enumerate(best_traj):
        plt.plot(traj[:, 0], traj[:, 1], lw=2, alpha=0.8, label=f"Traj {i+1}")
    plt.scatter(state[0], state[1], color='red', s=60, label="Start", zorder=5)
    plt.axvline(0.5, color='gray', ls='--', label='Goal position')
    plt.xlabel("Position")
    plt.ylabel("Velocity")
    plt.title(f"Predicted Trajectories (Continuous MPC)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Finding optimal action for MPC (Random Shooting)

```python
import matplotlib.pyplot as plt
import numpy as np

# --------------------------------------------------------------
#  MPC (Random Shooting)
# --------------------------------------------------------------

def mpc_action(state, H=25, N_seq=200, plot=False, plot_best=10, repeat=5):
    """
    Random Shooting MPC for continuous actions in [-1, 1].
    Each GP predicts Œîposition and Œîvelocity given (pos, vel, action).
    """

    ########### TODO ####################
    # 1. Sample action sequences
    U = np.random.uniform(low=-1.0, high=1.0, size=(N_seq, H))

    # 2. Roll out trajectories
    all_traj, J = roll_out_n_trajectories(state, U)

    # 3. Select top trajectories
    best_idx = np.argsort(J)[:min(plot_best, N_seq)]
    best_seq_idx = best_idx[0]

    # Plot the best predicted trajectories
    if plot:
        best_traj = [all_traj[i] for i in best_idx]
        plot_trajectories(state, best_traj)

    # 4. Return the first action of the best sequence
    best_action = U[best_seq_idx, 0]
    return best_action


obs, _ = env.reset(seed=0)
a = mpc_action(obs, H=80, N_seq=100, plot=True, plot_best=4)
```

```python
# --------------------------------------------------------------
# Running in the gym environment
# --------------------------------------------------------------
H = 80        # MPC horizon
N_seq = 200   # Number of sampled action sequences
steps = 500   # Max episode length


obs, _ = env.reset()
returns = 0.0
a = 0.0       # initial action

trajectory = []

for t in range(steps):
    # --- 1. Compute optimal action via MPC ---
    a = mpc_action(obs, H=H, N_seq=N_seq)

    if t%10==0:
        print("t=",t, "state=", obs, "a=",a)

    # --- 2. Step environment ---
    next_obs, r, term, trunc, _ = env.step([a])  # continuous env expects array-like
    trajectory.append([obs,a])
    env.render()  # draw each step

    returns += r

    # --- 5. Prepare next step ---
    obs = next_obs
    if term or trunc:
        break

print(f"Episode return: {returns:.2f}")
env.close()
```

```python
# trajectory is a list of [obs, a]
# each obs is [pos, vel]

def plot_trajectory(trajectory):
    positions, velocities, actions = [], [], []

    for entry in trajectory:
        obs, a = entry
        positions.append(obs[0])
        velocities.append(obs[1])
        actions.append(float(a))  # ensure scalar

    positions = np.array(positions)
    velocities = np.array(velocities)
    actions = np.array(actions)

    # --------------------------------------------------------------
    # Scatter plot: position vs velocity, colored by action
    # --------------------------------------------------------------
    plt.figure(figsize=(8, 6))
    sc = plt.scatter(positions, velocities, c=actions,
                     cmap='coolwarm', s=50, edgecolor='k', alpha=0.8)

    cbar = plt.colorbar(sc, label="Action (acceleration)")
    plt.xlabel("Position")
    plt.ylabel("Velocity")
    plt.title("MountainCar Trajectory ‚Äî Colored by Action")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_trajectory(trajectory)
```

# üß© Exercise 2: Implement the MPPI Controller

In this exercise, you will complete the **`action()` method** of the `MPPIController` class.  
This method implements the **Model Predictive Path Integral (MPPI)** algorithm, which selects an optimal control action based on sampled action sequences and their simulated costs.

---

## ‚öôÔ∏è MPPI Overview

Model Predictive Path Integral (MPPI) control is a **sampling-based model predictive controller**.  
It explores many possible future action sequences and evaluates them under a dynamics model.  
Unlike random shooting, MPPI adds **Gaussian noise around a mean sequence** and keeps track of a **persistent best sequence** that shifts at every step.

At each time step:

1. **Shift** the previous mean action sequence (`U_mean`) forward in time.  
   ‚Üí This reuses information from the last solution.

2. **Sample** new action sequences by adding Gaussian noise around this mean.

3. **Roll out** each sequence using the system dynamics to compute its total cost.

4. Mean computation:
   * For undergraduate: **Select** the best trajectory (the one with the lowest cost) as the mean. (MPPI Simplified version)
   * For graduate: Compute **importance-weighted averages of control perturbations**, following the path-integral formulation of stochastic optimal control.

6. **Return** the first action of the best sequence to apply to the system.

# MPPI

```python
import numpy as np
import matplotlib.pyplot as plt

class MPPIController:
    def __init__(self, H=100, N_seq=200, Œª=1.0, œÉ=0.3):
        """
        Model Predictive Path Integral Controller with persistent action sequence.
        Uses the best trajectory (lowest-cost) to update instead of weighted mean.
        """
        self.H = H
        self.N_seq = N_seq
        self.Œª = Œª
        self.œÉ = œÉ


        # Persistent mean sequence for warm-start
        self.U_mean = np.zeros(H)

    def action(self, state, plot=False, plot_best=3):
        H, N_seq, œÉ = self.H, self.N_seq, self.œÉ

        ######## TODO #############
        # 0 Shift previous mean sequence (reuse previous solution)
        self.U_mean[:-1] = self.U_mean[1:]
        self.U_mean[-1] = 0.0

        # 1. Sample action sequences (Gaussian noise around previous U_mean)
        U = self.U_mean[None, :] + œÉ * np.random.randn(N_seq, H)
        U = np.clip(U, -1.0, 1.0)

        # 2. Roll out trajectories
        all_traj, J = roll_out_n_trajectories(state, U)

        # 3. Select the best trajectory (lowest cost)
        best_idx = int(np.argmin(J))
        self.U_mean = U[best_idx].copy()

        #  Optional: plot best trajectories
        if plot:
            best_k = np.argsort(J)[:plot_best]
            best_traj = [all_traj[i] for i in best_k]
            plot_trajectories(state, best_traj)

        # 4. Return the first action of the best sequence
        best_action = self.U_mean[0]
        return best_action

# Initialize controller once
mppi = MPPIController(H=80, N_seq=150, Œª=1.0, œÉ=2.1)

obs, _ = env.reset()
a = mppi.action(obs, plot=True, plot_best=3)  # plot first iteration only
a
```

```python
# ---------- Reward / Cost: Stop-at-goal (a) ----------
def reward_stop_at_goal(state, a, goal=0.5, w_p=200.0, w_v=50.0, w_u=0.1):
    x, v = state
    # Î™©Ìëú Í∑ºÏ≤òÏóê ÏûàÏùÑÏàòÎ°ù(ÏúÑÏπò Ïò§Ï∞®‚Üì) + ÏÜçÎèÑÍ∞Ä ÏûëÏùÑÏàòÎ°ù(Ï†ïÏßÄÏóê Í∞ÄÍπùÍ≤å) + Ï†úÏñ¥ ÎÖ∏Î†•ÏùÄ ÏûëÍ≤å
    # Î≥¥ÏÉÅÏùÄ ÌÅ∞Í∞íÏù¥ Ï¢ãÏùå ‚Üí ÏúÑÏπò/ÏÜçÎèÑÎäî ÏùåÏùò Ï†úÍ≥±Ìï≠, Ï†úÏñ¥ÎèÑ ÏùåÏùò Ï†úÍ≥±Ìï≠
    return -(w_p * (x - goal)**2 + w_v * (v**2) + w_u * (a**2))

def cost_stop_at_goal(state, a, goal=0.5, w_p=200.0, w_v=50.0, w_u=0.1):
    return -reward_stop_at_goal(state, a, goal, w_p, w_v, w_u)


# ---------- Reward / Cost: Energy-saving (b) ----------
def reward_energy_saving(state, a, goal=0.5, w_p=200.0, w_v=50.0, w_u=3.0):
    x, v = state
    # (a)Ïùò Íµ¨Ï°∞Îäî Ïú†ÏßÄÌïòÎêò, Ï†úÏñ¥ ÎπÑÏö© Í∞ÄÏ§ëÏπò w_u ‚Üë (ÏóêÎÑàÏßÄ Ï†àÏïΩ Í∞ïÌôî)
    return -(w_p * (x - goal)**2 + w_v * (v**2) + w_u * (a**2))

def cost_energy_saving(state, a, goal=0.5, w_p=200.0, w_v=50.0, w_u=3.0):
    return -reward_energy_saving(state, a, goal, w_p, w_v, w_u)
```

```python
def roll_out_trajectory_with_cost(s0, Ui, cost_fun, gamma=1.0):
    s = s0.copy()
    traj = [s0.copy()]
    H = len(Ui)
    discount = 1.0
    total_cost = 0.0

    for k in range(H):
        a = Ui[k]
        s = mountain_car_dynamics(s[0], s[1], a)
        traj.append(s.copy())
        total_cost += cost_fun(s, a) * discount
        discount *= gamma
    return np.array(traj), total_cost


def roll_out_n_trajectories_with_cost(state, U, cost_fun):
    N_seq = len(U)
    J = np.zeros(N_seq)
    all_traj = []
    for i in range(N_seq):
        traj, total_cost = roll_out_trajectory_with_cost(state, U[i], cost_fun)
        J[i] = total_cost
        all_traj.append(np.array(traj))
    return all_traj, J
```

```python
def mpc_action(state, H=25, N_seq=200, plot=False, plot_best=10, repeat=5, cost_fun=cost_stop_at_goal):
    # 1) sample
    U = np.random.uniform(low=-1.0, high=1.0, size=(N_seq, H))

    # 2) roll out with chosen cost
    all_traj, J = roll_out_n_trajectories_with_cost(state, U, cost_fun)

    # 3) select
    best_idx = np.argsort(J)[:min(plot_best, N_seq)]
    best_seq_idx = best_idx[0]

    # plot
    if plot:
        best_traj = [all_traj[i] for i in best_idx]
        plot_trajectories(state, best_traj)

    # 4) act
    best_action = U[best_seq_idx, 0]
    return np.array([best_action])
```

```python
class MPPIController:
    def __init__(self, H=100, N_seq=200, Œª=1.0, œÉ=0.3, cost_fun=cost_stop_at_goal):
        self.H = H
        self.N_seq = N_seq
        self.Œª = Œª
        self.œÉ = œÉ
        self.cost_fun = cost_fun
        self.U_mean = np.zeros(H)

    def action(self, state, plot=False, plot_best=3):
        H, N_seq, œÉ = self.H, self.N_seq, self.œÉ

        # shift warm-start
        self.U_mean[:-1] = self.U_mean[1:]
        self.U_mean[-1] = 0.0

        # sample around mean
        U = self.U_mean[None, :] + œÉ * np.random.randn(N_seq, H)
        U = np.clip(U, -1.0, 1.0)

        # roll out with chosen cost
        all_traj, J = roll_out_n_trajectories_with_cost(state, U, self.cost_fun)

        # pick best; update mean
        best_idx = int(np.argmin(J))
        self.U_mean = U[best_idx].copy()

        if plot:
            best_k = np.argsort(J)[:min(plot_best, N_seq)]
            best_traj = [all_traj[i] for i in best_k]
            plot_trajectories(state, best_traj)

        return np.array([self.U_mean[0]])
```

```python
def run_episode(controller, env, steps=300, render=False):
    obs, _ = env.reset()
    xs, vs, us = [], [], []
    for t in range(steps):
        if render:
            env.render()
        a = controller.action(obs, plot=False) if hasattr(controller, 'action') else controller(obs)
        # Gymnasium Continuous MountainCar expects shape (1,)
        obs, r, terminated, truncated, info = env.step(a)
        xs.append(obs[0]); vs.append(obs[1]); us.append(float(a[0]))
        if terminated or truncated:
            break
    return np.array(xs), np.array(vs), np.array(us)

def plot_xvu(xs, vs, us, title=""):
    T = np.arange(len(xs))
    plt.figure(figsize=(10, 7))
    plt.subplot(3,1,1); plt.plot(T, xs); plt.axhline(0.5, ls='--', c='gray'); plt.ylabel("position x")
    plt.subplot(3,1,2); plt.plot(T, vs); plt.ylabel("velocity v")
    plt.subplot(3,1,3); plt.plot(T, us); plt.ylabel("action u"); plt.xlabel("time step")
    plt.suptitle(title); plt.tight_layout(); plt.show()
```

```python
# (a) Stop-at-goal
mppi_a = MPPIController(H=80, N_seq=150, Œª=1.0, œÉ=2.1, cost_fun=cost_stop_at_goal)
xs_a, vs_a, us_a = run_episode(mppi_a, env, steps=400)
plot_xvu(xs_a, vs_a, us_a, title="(a) Stop-at-goal: position/velocity/action")

# (b) Energy-saving
#   ‚Üí Í∞ôÏùÄ Íµ¨Ï°∞ÏßÄÎßå Ï†úÏñ¥ Í∞ÄÏ§ëÏπòÍ∞Ä ÌÅ∞ cost_energy_saving ÏÇ¨Ïö©
mppi_b = MPPIController(H=80, N_seq=150, Œª=1.0, œÉ=2.1, cost_fun=cost_energy_saving)
xs_b, vs_b, us_b = run_episode(mppi_b, env, steps=400)
plot_xvu(xs_b, vs_b, us_b, title="(b) Energy-saving: position/velocity/action")
```

(a) Stop-at-Goal Reward

The redesigned reward encourages the car to reach the goal position and stop smoothly once it gets there. By penalizing both the distance to the goal and the velocity magnitude, the controller learns to slow down as it approaches the hilltop rather than overshooting it. The additional penalty on the control effort prevents unnecessary oscillations near the goal. In the resulting trajectories, the car accelerates strongly at first, then gradually reduces its speed and finally settles around x ‚âà 0.5 with near-zero velocity and small control inputs, showing stable convergence.

(b) Energy-Saving Reward

In the energy-saving version, the cost weight on the control term a¬≤ was increased to emphasize energy efficiency. This makes the controller use smaller thrusts throughout the episode. As a consequence, the car climbs the hill more slowly and smoothly, exhibiting lower peak velocity and weaker control amplitudes. Although the goal is reached later, the motion is gentler and consumes less energy overall. The stronger penalty on control effort therefore trades off speed for efficiency, yielding smoother but slower convergence to the target position.
